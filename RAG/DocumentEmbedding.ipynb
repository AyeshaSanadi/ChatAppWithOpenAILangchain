Indexing: Text Embedding with OpenAI

```python
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters.markdown import MarkdownHeaderTextSplitter
from langchain_text_splitters.character import CharacterTextSplitter
from langchain_openai.embeddings import OpenAIEmbeddings
import numpy as np
```

```python
docs_loader = Docx2txtLoader("Introduction_to_Data_and_Data_Science2.docx")
pages = docs_loader.load()

md_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=[("#","Course title"), ("##", "Lecture Title")])
md_pages = md_splitter.split_text(pages[0].page_content)

for i in range(len(md_pages)):
    md_pages[i].page_content = " ".join(md_pages[i].page_content.split())

char_split = CharacterTextSplitter(separator=".", chunk_size=500, chunk_overlap=50)

pages_char_split = char_split.split_documents(md_pages)
```

```python
pages_char_split
```

```python
embedding = OpenAIEmbeddings(model = "text-embedding-ada-002")
```

```python
vector1 = embedding.embed_query(pages_char_split[1].page_content)
vector2 = embedding.embed_query(pages_char_split[9].page_content)
vector3 = embedding.embed_query(pages_char_split[15].page_content)
```

```python
vector1
```

```python
len(vector1), len(vector2), len(vector3)
```

```python
# apply the dot product to each pair of vectors and quantify their similarity
np.dot(vector1, vector2), np.dot(vector2, vector3), np.dot(vector3, vector1)
```
